# **Jogo - Pedra, Papel e Tesoura**

```python
print("------JOGO PEDRA, PAPEL E TESOURA------------------")
print("\n")

jogar = "SIM"
opcoes_validas = ("PEDRA", "PAPEL", "TESOURA")

print(f"Opções válidas: {opcoes_validas}")

while jogar == "SIM" or jogar == "S":
  jogador1 = input("Jogador 1, digite sua opção: ").upper().strip()
  if jogador1 not in opcoes_validas:
    print("Opção inválida")
    break
  jogador2 = input("Jogador 2, digite sua opção: ").upper().strip()
  if jogador2 not in opcoes_validas:
      print("Opção inválida")
      break

  if jogador1 == jogador2:
    print("Empate!!")

  elif ((jogador1 == "PEDRA" and jogador2 == "TESOURA") or
    (jogador1 == "PAPEL" and jogador2 == "PEDRA") or
    (jogador1 == "TESOURA" and jogador2 == "PAPEL")):
    print("Jogador 1, venceu!!")

  else:
    print("Jogador 2, venceu!!")

  jogar = input("Deseja jogar novamente? ").upper()

```
