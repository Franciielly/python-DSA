**Variável:** Espaço na memória destinado ao armazenamento de dados.


Python:
*   Dinâmico
*   Fortemente tipado
*   Case sensitive

**Variáveis Global:** são definidas fora de qualquer função e podem ser acessadas em qualquer parte do código.

**Variáveis Locais:** são definidas dentro de uma função e só podem ser acessadas dentro dessa função.
Elas deixam de existir quando a função termina.



# **Tipos Primitivos de Dados**

esse são os tipos mais básicos em Python:

```python
#Integer (Inteiro)
n_inteiro = 100
print(f"Valor: {n_inteiro}, Tipo{type(n_inteiro)}")
```

```python
#Float (Ponto flutuante)
n_float = 10.50
print(f"Valor: {n_float}, Tipo: {type(n_float)}")
```

```python
#String (Texto)
texto = "Python é a melhor linguagem"
print(f"Valor: {texto}, Tipo: {type(texto)}")
```

```python
#Boolean (Boleano)
verdadeiro = True
falso = False

print(f"Valor: {verdadeiro}, Tipo: {type(verdadeiro)}")
print(f"Valor: {falso}, Tipo: {type(falso)}")
```


# **Manipulação de Strings**

```python
frase = " Aprender Python é legal!! "
```

```python
#Função len - tamanho da string
print(f"Tamanho da frase: {len(frase)}")
```

```python
#Função upper e lower - maiúscula e minúscula
print(frase.upper())
print(frase.lower())
```

```python
#Função strip - remover espaços em brancos
frase_nova = frase.strip()
print(f"Frase sem espaços: {frase_nova}")
print(len(frase_nova))
```

```python
#Função replace - substituir textos em texto de execução
print(f"Substituindo 'legal' por 'divertido': {frase_nova.replace('legal', 'divertido')} ")
```

```python
#Função Slicing - fatiamento
print(frase_nova)
print(f"O primeiro caractere:{frase_nova[0]}")
print(f"Palvra python: {frase_nova[9:15]}")
```

# **Estrutura de Dados**

*  **Lista**: coleção ordenadas e **mutáveis**. Pode conter diferentes tipos de dados






```python
#Criando lista
frutas = ["Maçã", "Laranja", "Uva", "Melancia"]
print (f"Lista de frutas: {frutas}")
```

```python
type(frutas)
```

```python
#acessando fruta por índice
print(f"A primeira fruta é: {frutas[0]}")
print(f"A última fruta é: {frutas[-1]}")
```

```python
#Adicionando itens
fruta = input("Digite o nome da fruta: ")
frutas.append(fruta)
print(f"Lista atalizada: {frutas}")
```

```python
#Remover item
frutas.remove("Uva")
print(f"Lista atualizada: {frutas}")
```

```python
#Modificando lista
frutas[0] = "Morango"
print(f"Lista atualizada: {frutas}")
```

```python
#Verificando tamanho da lista
print(f"A lista tem {len(frutas)} frutas")
```

```python
#Deletando lista
del frutas
```



*   **Tuplas:** coleção ordenadas e imutáveis


```python
#Criando tupla
coordenadas = (10.2, 20.5)
print(f"Tupla de coordenadas: {coordenadas}")
```

```python
type(coordenadas)
```

```python
#Acessando uma coordenada pelo índice
print(f"Coordenada X: {coordenadas[0]}")
print(f"Coordenada Y: {coordenadas[1]}")
```

```python
#Tuplas são úteis para dados que não vamos mudar, como dia da semana, meses do ano, etc
dia_semana = ("Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo")
print(f"O primeiro dia da semana é: {dia_semana[0]}")
```

```python
#Transformando Tupla em Lista
# coordenadasD = [coordenadas] - lista contendo tupla
coordenadasD = list(coordenadas)
type(coordenadasD)
```



*   **Dicionário:** coleção chave: valor. São mutáveis



```python
#Criando dicionário
aluno = {
    "nome": "Bob",
    "idade": 22,
    "curso": "Ciência de Dados",
    "aluno_ativo": True
}
print(f"Dicionário do aluno: {aluno}")
```

```python
type(aluno)
```

```python
#Acessando valor pela chave
print(f"Nome: {aluno['nome']}")
print(f"Curso: {aluno.get('curso')}")  # .get() é uma forma segura de acessar os dados
```

```python
#Adicionando par chave
aluno ["Cidade"] = "São Paulo"
print(f"Dicionário atualizado: \n {aluno}")
```

```python
#Modificando dicionário
aluno["idade"] = 23
print(f"Dicionário modificado: {aluno['idade']}")
print(f"Dicionário modificado: {aluno}")
```

```python
#Removendo um par chave
del aluno["aluno_ativo"]
print(f"Dicionário após remover aluno ativo: {aluno}")
```


*   **Conjuntos(sets):** coleção não ordenadas, itens únicos e mutáveis. Utíl para remover duplicatas

```python
#Criando um conjunto
numeros = {1, 2, 3, 4, 3, 2, 5}
print(f"Conjunto de números (sem duplicatas): {numeros}")
```

```python
type(numeros)
```

```python
#Adicionando um item
numeros.add(6)
print(f"Após adicionar o valor 6: {numeros}")
```

```python
#Removendo um item
numeros.remove(2)
print(f"Após remover o 2: {numeros}")
```

```python
#Operações de conjuntos
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6}
```

```python
#União
conjuntos = conjunto_a.union(conjunto_b)
print(f"União de conjuntos: {conjuntos}")
```

```python
#Interseção
conjuntosI = conjunto_a.intersection(conjunto_b)
print(f"Interseção de conjuntos: {conjuntosI}")
```

# **Conversão de Dados (Type casting)**
conversão de um tipo de dados para outro

```python
#String -> inteiro
numeroS = "123"
numeroI = int(numeroS)
print(f" Número em string: {numeroS} {type(numeroS)}\n Número em Int: {numeroI} {type(numeroI)}")
```

```python
#Convertendo Estrura de Dados
lista_duplicatas = [1, 2, 3, 4, 4, 4, 5, 6, 2]
conjunto_unico = set(lista_duplicatas)
lista_sem_duplicatas = list(conjunto_unico)

print(f"Lista original: {lista_duplicatas}")
print(f"Conjunto: {conjunto_unico}")
print(f"Lista sem duplicatas: {lista_sem_duplicatas}")
```
