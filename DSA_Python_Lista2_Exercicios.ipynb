#### Exercício 1

Escreva uma função que receba os valores de três lados de um triângulo e o classifique como "Equilátero" (todos os lados iguais), "Isósceles" (dois lados iguais) ou "Escaleno" (todos os lados diferentes).

```python
def triangulo(lado1, lado2, lado3):

  #Verifica se forma um triângulo
  if (lado1 + lado2 > lado3 and
        lado1 + lado3 > lado2 and
        lado2 + lado3 > lado1):

    if lado1 == lado2 == lado3:
      return "Triângulo Equilátero"
    elif lado1 == lado2 or lado2 == lado3 or lado1 == lado3:
      return "Triângulo Isósceles"
    else:
      return "Triângulo Escaleno"

  else:
    return "Não forma um triângulo"


lado1 = float(input("Digite o lado 1 do triângulo: "))
lado2 = float(input("Digite o lado 2 do triângulo: "))
lado3 = float(input("Digite o lado 3 do triângulo: "))

resultado = triangulo(lado1, lado2, lado3)

print(f"{resultado}.")
```

#### Exercício 2

Escreva uma função que recebe um número inteiro e exibe a tabuada de multiplicação desse número, do 1 ao 10.

```python
def tabuada():
  numero = int(input("Digite um número inteiro: "))

  print(f"\n--------------TABUADA DO {numero}---------------")
  for n in range(1, 11, 1):
    print(f"{numero} X {n} = {numero*n}")


tabuada()
```

#### Exercício 3

Você recebeu um dicionário com os nomes dos alunos e suas respectivas notas. Escreva uma função que calcula a média da turma e retorna uma lista com os nomes dos alunos que tiveram nota acima da média.

```python
def acimaMedia(dicionario):
  media = sum(dicionario.values()) / len(dicionario)

  aluno_acima_media = []

  for aluno, nota in dicionario.items():
    if nota > media:
      aluno_acima_media.append(aluno)

  return media, aluno_acima_media

dicionario = {
    "Ana": 8.5,
    "Bruno": 6.0,
    "Carlos": 7.5,
    "Daniela": 9.0,
    "Eduardo": 5.5,
    "Fernanda": 8.0,
    "Gabriel": 6.5,
    "Helena": 9.5,
    "Igor": 7.0,
    "Juliana": 8.8,
    "Lucas": 4.9,
    "Mariana": 7.9,
    "Nathan": 6.8,
    "Olivia": 9.2,
    "Pedro": 5.7
}

media, lista = acimaMedia(dicionario)

print(f"A média da turma é de {media:.2f}")
print(f"Alunos acima da média: {lista}")
```

#### Exercício 5

Crie uma função chamada dsa_calcula_imc que aceite dois argumentos: peso (em kg) e altura (em metros). A função deve calcular o Índice de Massa Corporal (IMC) usando a fórmula IMC=peso/altura^2 e retornar o valor do IMC.

```python
def dsa_calcula_imc(peso, altura):
  imc = peso/(altura ** 2)

  if imc < 18.5:
    return f"Você está muito abaixo do peso. Imc: {imc}"
  elif imc < 20:
    return f"Você está magro. Imc: {imc:.2f}"
  elif imc < 24.9:
    return f"Você está no peso ideal. Imc: {imc:.2f}"
  elif imc < 29.9:
    return f"Você está com sobrepeso. Imc: {imc:.2f}"
  else:
    return f"Você está com obesidade. Imc: {imc:.2f}"


peso = float(input("Digite seu peso(kg): "))
altura = float(input("Digite sua altura: "))

print(dsa_calcula_imc(peso, altura))
```

#### Exercício 7

Crie uma função que receba uma lista de números inteiros e retorne um dicionário contendo a contagem de quantos números são pares e quantos são ímpares.

```python
def par_impar(lista):
  resultado = {
      "pares":0,
      "impares": 0
  }
  for n in lista:
    if n % 2 ==0:
      resultado["pares"] += 1
    else:
      resultado["impares"] += 1
  return resultado

lista = [1,2,3,4,45,65,47]
print(f"Quantidade de pares e impares na lista{lista}\n {par_impar(lista)}")
```

#### Exercício 8

Crie uma função que receba uma lista de strings (potenciais e-mails) e um parâmetro opcional dominio_desejado (com valor padrão "gmail.com"). A função deve retornar uma nova lista, usando list comprehension, contendo apenas os e-mails que terminam com o domínio desejado.

```python
def filtrar_email(emails, dominio_desejado = "@gmail.com"):
  return [email
          for email in emails
          if email.endswith(dominio_desejado)]

lista_emails = [
    "ana@gmail.com",
    "joao@yahoo.com",
    "maria@gmail.com",
    "carlos@hotmail.com"
]

resultado = filtrar_email(lista_emails)
print(resultado)
```

#### Exercício 9

Dada uma lista de frases, use a função map() em conjunto com uma expressão lambda para criar uma nova lista onde cada frase é convertida para letras maiúsculas e tem a palavra "PYTHON" anexada ao final.

```python
def transformar_frase(lista_frase):
  novas_frases = list(
      map(lambda frase : frase.upper() + " PYTHON", lista_frase)
  )
  return novas_frases

frases = [
    "eu gosto de programar",
    "estou aprendendo python",
    "funções são importantes"
]

resultado = transformar_frase(frases)
print(resultado)
```

#### Exercício 10

Crie um jogo onde o computador escolhe um número secreto entre 1 e 20. O jogador tem 5 tentativas para adivinhar. A cada tentativa, o programa informa se o palpite foi muito alto ou muito baixo. Se o jogador acertar, o loop deve ser interrompido com uma mensagem de vitória.

```python
import random

numero = random.randint(1, 20)
acertou = False

print("=" * 50)
print("              JOGO ADIVINHAÇÃO")
print("=" * 50)
for i in range(5):
  tentativa = int(input("Digite um número inteiro entre 1 a 20: "))
  if tentativa < 1 or tentativa > 20:
    print("Digite um número entre 1 e 20\n")
    continue

  if tentativa == numero:
    print("Parabéns, você acertou o número secreto!!!")
    acertou = True
    break

  tentativas_restantes = 5 - (i + 1)

  if tentativa < numero:
    print(f"Palpite baixo. Você tem {tentativas_restantes} tentativas\n")
  else:
    print(f"Palpite alto. Você tem {tentativas_restantes} tentativas\n")

if not acertou:
  print(f"Infelizmente você não descobriu. O número secreto era {numero}.")
```
